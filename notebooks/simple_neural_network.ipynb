```python
# Importieren der benötigten Module
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

**Datenvorbereitung**

```python
train_labels = []
train_samples = []
```

Bei 1000 Personen wird ein neues Medikament getestet. 50% der Personen sind über 65 Jahre alt bzw. jünger als 65 Jahre alt. 95% der über 65 Jahre alten Personen haben Nebenwirkungen. 5% der jünger als 65 Jahre alten Personen haben Nebenwirkungen.

```python
for i in range(50):
        # 5% der jungen Personen mit Nebenwirkungen
        random_younger = randint(13,64)
        train_samples.append(random_younger)
        train_labels.append(1)
        # 5% der alten Personen ohne Nebenwirkungen
        random_older = randint(65,100)
        train_samples.append(random_older)
        train_labels.append(0)
        
for i in range(950):
        # 95% der jungen Personen ohne Nebenwirkungen
        random_younger = randint(13,64)
        train_samples.append(random_younger)
        train_labels.append(0)
        # 95% der alten Personen mit Nebenwirkungen
        random_older = randint(65,100)
        train_samples.append(random_older)
        train_labels.append(1)
```

```python
print(len(train_samples))
```

```python
for i in train_samples:
    print(i)
```

```python
for i in train_labels:
    print(i)
```

```python
# Daten werden in Numpy arrays umgewandelt und gemischt.
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
train_labels, train_samples = shuffle(train_labels, train_samples)
```

```python
# Daten werden zu Werten zwischen 0 und 1 skaliert
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

**Erstellung des NN**

```python
# Tensorflow Module importieren
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
# Lässt die Berechnungen auf der GPU laufen
physical_devices = tf.config.experimental.list_physical_devices('GPU')
print("Num GPUs available:", len(physical_devices))
# tf.config.experimental.set_memory_growth(physical_devices[0], True)

#Nachlesen, Nummer der genutzten CPUs ausgeben lassen.
```

```python
# Neural Network mit 2 hidden layers erstellen
model = Sequential([
    Dense(units=128, input_shape=(1,), activation='relu'),
    Dense(units=2, activation='softmax')    
])
```

```python
model.summary()
```

```python
# Erstellung des Modells mit lr und loss_function
model.compile(optimizer=Adam(learning_rate=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Trainieren des NN 
model.fit(x=scaled_train_samples, y=train_labels, validation_split=0.1, batch_size=10, epochs=30, shuffle=True, verbose=2)
```

**Vorbereitung der Testdaten**

```python
test_labels = []
test_samples = []
```

```python
for i in range(10):
        # 5% der jungen Personen mit Nebenwirkungen
        random_younger = randint(13,64)
        test_samples.append(random_younger)
        test_labels.append(1)
        # 5% der alten Personen ohne Nebenwirkungen
        random_older = randint(65,100)
        test_samples.append(random_older)
        test_labels.append(0)
        
for i in range(190):
        # 95% der jungen Personen ohne Nebenwirkungen
        random_younger = randint(13,64)
        test_samples.append(random_younger)
        test_labels.append(0)
        # 95% der alten Personen mit Nebenwirkungen
        random_older = randint(65,100)
        test_samples.append(random_older)
        test_labels.append(1)
```

```python
test_labels = np.array(test_labels)
test_samples = np.array(test_samples)
test_labels, test_samples = shuffle(test_labels, test_samples)
```

**Cofusion Matrix**

```python
%matplotlib inline
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(y_true=test_labels, y_pred=rounded_predictions)
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion Matrix',
                          cmap=plt.cm.Blues):
    """
    Diese Funktion printet und plottet die Confusion Matrix.nm 
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print("Confusion matrix, without normalization")
        
    print(cm)
    
    thresh = cm.max() / 2
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
    
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
cm_plot_labels = ['no_side_effects', 'had_side_effects']
plot_confusion_matrix(cm=cm, classes=cm_plot_labels, title='Confusion Matrix')
```
