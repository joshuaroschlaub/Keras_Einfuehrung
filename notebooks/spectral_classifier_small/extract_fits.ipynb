```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
from astropy.wcs import WCS
import os
import csv
```

```python
fits_path = 'F:\\data\\spectral_fits\\'
csv_path = 'F:\\data\\spectral_csv\\'

for directory in os.listdir(fits_path):
    
    count_failed=0
    count_added=0
    
    for filename in os.listdir(fits_path + directory + '\\'):
        
        path = fits_path + '\\' + directory + '\\' + filename
        
        # fits-Dateien öffnen und wavelength + flux einlesen
        hdul = fits.open(path)
        data = hdul[1].data
        flux = data['flux']
        wavelength = 10**data['loglam']
        hdul.close()
        
        # Find first und last Index finden
        for i in range(len(wavelength)):
            if wavelength[i]>4000:
                first_index = i
                break

        for i in range(len(wavelength)):
            if wavelength[i]>9000:
                last_index = i
                break

        # wavenlength und flux Listen schneiden
        wavelength = wavelength[first_index:last_index]
        flux = flux[first_index:last_index]
        
        # Daten in csv-Dateien abspeichern
        if len(wavelength) == 3522 and count_added < 1000:
            file = open(csv_path + directory + '\\' + filename[:-5] + '.csv', 'w') 
            with file:
                writer = csv.writer(file, delimiter=',', lineterminator='\n')
                writer.writerow(["wavelength","flux"])

                for i in range(len(wavelength)):
                    writer.writerow([wavelength[i],flux[i]])
            count_added += 1
                    
        if len(wavelength) != 3522:
            print("Länge der Liste wavelength ist: " + str(len(wavelength)))
            count_failed += 1
            
    print(str(count_failed/1100*100) + "% waren nicht erfolgreich bei der Klasse:" + directory)
    print(str(count_added) + " CSV-Dateien angelegt.")
```
